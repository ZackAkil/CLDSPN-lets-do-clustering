```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## Generate clusterable data
Modify the `data_spread` variable to see it effect the data

```python
data_spread = 1

data = make_blobs(random_state=42,
                  centers=3, 
                  cluster_std=data_spread, 
                  n_samples=100)[0]

plt.scatter(data[:,0], data[:,1])
plt.show()
```

## Perform elbow analysis
This just involves running kmean with more and more clusters and loning the `inertia_`.

The `inertia_` is the sum of the distances between data and their centroids.

```python
distances = []

for i in range(1,10):
    clst = KMeans(n_clusters=i)
    clst.fit(data)
    distances.append(clst.inertia_)
    
    # un-comment the following code to plot out every kmeans model
    
#     plt.scatter(data[:,0], data[:,1], c=clst.predict(data))
#     plt.show()
```

##  plot the elbow

```python
plt.plot(list(range(1,10)), distances)
plt.show()
```
